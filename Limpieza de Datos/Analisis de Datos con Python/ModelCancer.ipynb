```python
# Rango Intercuartílico (IQR)
from scipy import stats
data_sin_outliers=data.copy()
data_sin_outliers['NSP-Without Outlier'] = data_sin_outliers['Number of sexual partners']
# Calcular límites para identificar outliers
q1 = data_sin_outliers['NSP-Without Outlier'].quantile(0.25)
q3 = data_sin_outliers['NSP-Without Outlier'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Delete outliers
data_sin_outliers = data_sin_outliers[(data_sin_outliers['NSP-Without Outlier'] >= lower_bound) & (data_sin_outliers['NSP-Without Outlier'] <= upper_bound)]

```

```python
sns.set(rc={'figure.figsize':(15,12)})
sns.boxplot(data=data_sin_outliers,palette='rainbow',orient='h')
```
